## Nearby function returns offices that satisfy a preference in a radius

````python
def nearby_offices(lst,radius):
    res = []
    d = []
    for p in range(len(prefe)):
        for o in range(len(of)):      
            distance = gmaps.distance_matrix(of['office_str'][o],prefe[lst][p])['rows'][0]['elements'][0]['distance']['value']
            if distance < radius:
                res.append(of['office_str'][o])
                d.append(distance)
    return res,d
````

````python
def near_coordinate(lst,radius,choose):
    res = []
    d = []    
    for p in lst:
        for o in choose:
            distance = gmaps.distance_matrix(o,p)['rows'][0]['elements'][0]['distance']['value']
            if (distance < radius) and (o not in res):
                res.append(o)
                d.append(distance)
    return res,d
````
